>## **6. 🔥 Correlation Matrix (Heatmap)**  

Kita akan melihat korelasi antar variabel numerik dalam dataset menggunakan **Correlation Matrix**. Korelasi membantu kita memahami hubungan antara fitur-fitur yang ada.  

### 🛠 **Apa yang Akan Dilakukan?**  
✅ Menggunakan **matriks korelasi** untuk melihat hubungan antar variabel.  
✅ Menggunakan **heatmap dari Seaborn** untuk memvisualisasikan korelasi.  
✅ Menganalisis apakah ada variabel yang memiliki korelasi tinggi dengan **Depression**.  

```python
numerical_columns = df.select_dtypes(include=['number'])
correlation_matrix = numerical_columns.corr()
print(f"Matriks Korelasi: {correlation_matrix}")

plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix')
plt.show()

if 'Depression' in correlation_matrix.columns:
    print("\nKorelasi Variabel Lain dengan 'Depression':")
    correlation_with_depression = correlation_matrix['Depression'].drop('Depression', errors='ignore')
    print(correlation_with_depression.sort_values(ascending=False))
else:
    print("\nKolom 'Depression' tidak ditemukan dalam dataset.")
```

### Jawab insight yang kamu dapatkan dari visualisasi tadi!

warna merah artinya korelasi positif kuat, biru korelasi negative kuat. Berdasarkan heatmap, work pressure berkorelasi positif dengan depression, sedangkan job satisfaction dan sleep duration berkorelasi negarif, artinya tingkatan depresi berasal dari tekanan kerja yang tinggi, sementara kepuasan kerja dan durasi tidur dapat cukup untuk mengurangi depressi. CGPA berkorelasi positif dengan study satisfaction, hal ini menunjukkan bahwa mahasiswa yang memiliki nilai akademik yang bagus cenderung memiliki kepuasan terhadap cara

>## **7. 📊 Boxplot untuk Menganalisis Distribusi Data & Outlier**  

Boxplot membantu kita memahami **distribusi dan pencilan (outliers)** pada data numerik. Visualisasi ini sangat berguna untuk melihat persebaran data dan apakah ada nilai ekstrem dalam dataset.  

### 🛠 **Apa yang Akan Dilakukan?**

>## **8. 📊 Analisis Distribusi Kolom `Depression`**  

Salah satu analisis penting adalah melihat bagaimana distribusi data pada kolom **Depression** (apakah seseorang mengalami depresi atau tidak).  

#### 🛠 **Apa yang Harus Dilakukan?**  
✅ Buat **bar chart** atau **countplot** untuk melihat distribusi jumlah individu dengan dan tanpa depresi.  
✅ Identifikasi apakah dataset ini **seimbang (balanced)** atau **tidak seimbang (imbalanced)**.

```python
if 'Depression' in df.columns:
    plt.figure(figsize=(8, 6))
    sns.countplot(x='Depression', data=df, palette='coolwarm')
    plt.title("Distribusi Individu dengan dan tanpa Depresi")
    plt.xlabel("Status Depresi")
    plt.ylabel("Jumlah Individu")
    plt.show()

    distribution = df['Depression'].value_counts()
    print("Distribusi Jumlah Individu dengan dan tanpa Depresi:")
    print(distribution)

    total = len(df)
    for category, count in distribution.items():
        percentage = (count / total) * 100
        print(f"{category}: {count} individu ({percentage:.2f}%)")

    threshold = 10
    max_class = distribution.max()
    min_class = distribution.min()
    imbalance_percentage = ((max_class - min_class) / total) * 100

    if imbalance_percentage > threshold:
        print("\nDataset ini tidak seimbang (imbalanced).")
    else:
        print("\nDataset ini seimbang (balanced).")
else:
    print("Kolom 'Depression' tidak ditemukan dalam dataset.")
```

### Jawab insight yang kamu dapatkan dari visualisasi tadi!

Dari visualisasi distribusi kolom Depression, terlihat bahwa dataset ini tidak seimbang (imbalanced), dengan 81.83% individu tidak mengalami depresi, sedangkan 18.17% mengalami depresi. Ketidakseimbangan ini perlu diperhatikan dalam analisis lebih lanjut, terutama jika model machine learning digunakan, karena dapat mempengaruhi performa model. Jika ingin meningkatkan akurasi prediksi, teknik seperti oversampling, undersampling, atau penggunaan metrik evaluasi yang tepat dapat dipertimbangkan.

>## **10. 📊 Distribusi Gender**  

Melihat distribusi **gender** dapat membantu memahami komposisi peserta dalam dataset.  

#### 🛠 **Apa yang Harus Dilakukan?**  
✅ Gunakan **bar chart** atau **countplot** untuk melihat jumlah individu berdasarkan gender.  
✅ Cek apakah ada ketidakseimbangan dalam jumlah peserta laki-laki dan perempuan.

```python
if 'Gender' in df.columns:
    plt.figure(figsize=(8, 6))
    sns.countplot(x='Gender', data=df, palette='pastel')
    plt.title("Distribusi Individu Berdasarkan Gender")
    plt.xlabel("Gender")
    plt.ylabel("Jumlah Individu")
    plt.show()

    gender_distribution = df['Gender'].value_counts()
    print("Distribusi Jumlah Individu Berdasarkan Gender:")
    print(gender_distribution)
    total = len(df)
    for gender, count in gender_distribution.items():
        percentage = (count / total) * 100
        print(f"{gender}: {count} individu ({percentage:.2f}%)")

    if len(gender_distribution)==2:
      male_count = gender_distribution.iloc[0]
      female_count = gender_distribution.iloc[1]
      imbalance_percentage = abs((male_count - female_count) / total) * 100

      if imbalance_percentage > 10:
          print("\nDataset ini tidak seimbang dalam distribusi gender (imbalanced).")
      else:
          print("\nDataset ini seimbang dalam distribusi gender (balanced).")
else:
    print("Kolom 'Gender' tidak ditemukan dalam dataset.")
```

### Jawab insight yang kamu dapatkan dari visualisasi tadi!

Dari visualisasi distribusi gender, jumlah individu dalam dataset terbagi menjadi 55.06% laki-laki (77.464 individu) dan 44.94% perempuan (63.236 individu). Meskipun ada perbedaan, distribusi ini masih relatif seimbang, meskipun sedikit lebih banyak laki-laki. Namun, karena perbedaannya cukup signifikan (sekitar 10% lebih banyak laki-laki), dataset ini dianggap tidak sepenuhnya seimbang (imbalanced).

> ## **11. 🔎 Hubungan antara `Age` dan `Depression`**  

Apakah usia memiliki pengaruh terhadap kecenderungan mengalami depresi?
