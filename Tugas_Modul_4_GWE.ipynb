> ### üîç**3. Menangani Missing Values (Data Hilang)**    
- Tangani nilai yang hilang di kolom `Profession`, `Dietary Habits`, `Degree`, `Financial Stress`.  
- Bagaimana cara terbaik menangani data yang hilang ini? (Hapus, isi dengan rata-rata/median/modus, atau metode lain?)  
- Tangani nilai yang hilang di kolom `Academic Pressure`, `Work Pressure`, `Job Satisfaction`, `Study Satisfaction`

- Hintt :
- **Gabungkan `Academic Pressure` dan `Work Pressure`** menjadi satu kolom **`Pressure`**, dengan aturan:  
    - Jika individu adalah **Student**, gunakan nilai dari `Academic Pressure`.  
    - Jika individu adalah **Working Professional**, gunakan nilai dari `Work Pressure`.  
  - **Gabungkan `Study Satisfaction` dan `Job Satisfaction`** menjadi satu kolom **`Satisfaction`**, dengan aturan:  
    - Jika individu adalah **Student**, gunakan nilai dari `Study Satisfaction`.  
    - Jika individu adalah **Working Professional**, gunakan nilai dari `Job Satisfaction`.

- atau boleh juga nilai null di kolom `Academic Pressure`, `Work Pressure`, `Job Satisfaction`, `Study Satisfaction` nya diisi dengan 0

```python
### Answer Question 3 Here
import pandas as pd
df = pd.read_csv('dataset.csv')
# 1. Menangani nilai yang hilang di kolom kategori
for col in ['Profession', 'Dietary Habits', 'Degree']:
    df[col] = df[col].fillna(df[col].mode()[0])

# 2. Mengisi missing values untuk Financial Stress
df['Financial Stress'] = df['Financial Stress'].fillna(df['Financial Stress'].median())

# 3. Menggabungkan Academic Pressure dan Work Pressure menjadi satu kolom "Pressure"
df['Pressure'] = df.apply(lambda row: row['Academic Pressure'] if row['Working Professional or Student'] == 'Student'
                          else row['Work Pressure'], axis=1)

# Jika masih ada nilai kosong di "Pressure", isi dengan 0
df['Pressure'] = df['Pressure'].fillna(0)

# 4. Menggabungkan Study Satisfaction dan Job Satisfaction menjadi satu kolom "Satisfaction"
df['Satisfaction'] = df.apply(lambda row: row['Study Satisfaction'] if row['Working Professional or Student'] == 'Student'
                              else row['Job Satisfaction'], axis=1)

# Jika masih ada nilai kosong di "Satisfaction", isi dengan 0
df['Satisfaction'] = df['Satisfaction'].fillna(0)

# 5. Hapus kolom yang sudah digabung
df = df.drop(columns=['Academic Pressure', 'Work Pressure', 'Study Satisfaction', 'Job Satisfaction'])

# Menampilkan hasil setelah perbaikan missing values
df.info(), df.head()
```

>Berikan penjelasan pemilihan metode yang digunakan untuk mengisi nilai yang kosong disini :


Mengisi dengan modus untuk kolom kategori bukan numerik sehingga tidak bisa diisi dengan rata-rata atau median sehingga dengan modus dapat mengetahui nilai yang paling sering muncul untuk menampilkan mayoritas datanya.
Mengisi median untuk kolom numerik terhadap financial stress dapat menghindari outlier karena mengambil nilai tengah dari distribusi data.
Menggabungkan dua kolom dengan mengambil nilai sesuai tipe individu, Student menggunakan Academic Pressure dan Study Satisfaction, sementara Working Professional menggunakan Work Pressure dan job Satisfaction. Jika kosong, diisi 0.
Menghapus kolom lama karena adanya kolom yang sudah digabung atau data baru nya untuk mencegah redudansi atau menghindari adanya duplikasi data.


> ### üîç**4. Feature Engineering**  
-  Kolom `Sleep Duration` memiliki terlalu banyak unique value, kelompokkan kolom tersebut menjadi 3 kategori saja, **'Less than 7'**, **'7-8'**, **'More than 8'**
  
    

```python
### Answer Question 4 Here
def categorize_sleep(duration):
    if duration < 7:
        return "Less than 7"
    elif 7 <= duration <= 8:
        return "7-8"
    else:
        return "More than 8"

df['Sleep Duration'] = pd.to_numeric(df['Sleep Duration'], errors='coerce')

df['Sleep Category'] = df['Sleep Duration'].apply(categorize_sleep)

df['Sleep Category'].value_counts()
```

Untuk menghindari dummy variable trap, digunakan drop_first=True, yang menghapus satu kategori guna mengurangi redundansi dalam model.

  1. Identifikasi minimal **dua aspek preprocessing tambahan** yang diperlukan.

  3. Lakukan preprocessing yang sesuai dan dokumentasikan setiap langkahnya.

1. Identifikasi dua aspek preprocessing tambahan
Dua aspek preprocessing tambahan yang diperlukan:

- Penanganan Outlier

Outlier dapat menyebabkan model machine learning menjadi tidak stabil dan kurang akurat.
Solusi: Menggunakan Winsorizing untuk membatasi nilai ekstrem tanpa menghilangkan data.
- Encoding Kolom Kategori

Model machine learning hanya dapat bekerja dengan data numerik, sehingga perlu mengonversi kolom kategori (teks) menjadi angka.
Solusi: Menggunakan Label Encoding untuk mengganti teks dengan angka.

2. Dampak Preprocessing

- Winsorizing ‚Üí Mengurangi outlier, sehingga model lebih stabil.
- Mengisi Missing Values ‚Üí Mencegah error saat training model.
- Encoding Kolom Kategori ‚Üí Data bisa digunakan dalam model machine learning.
- PCA ‚Üí Mengurangi jumlah fitur tanpa kehilangan informasi penting.

```python
# Answer Question 9 Here
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import mstats  # Perbaikan import winsorize
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler, LabelEncoder

# 1. Load dataset (Pastikan dataset tersedia)
df = pd.read_csv("dataset.csv")  # Sesuaikan dengan nama file yang benar

# 2. Identifikasi dua aspek preprocessing tambahan
# - Outlier pada fitur numerik
# - Korelasi tinggi antar fitur

# 3. Menampilkan distribusi outlier dengan boxplot
plt.figure(figsize=(12, 6))
sns.boxplot(data=df.select_dtypes(include=[np.number]))
plt.xticks(rotation=45)
plt.title("Distribusi Fitur Numerik (Cek Outlier)")
plt.show()

# 4. Menampilkan heatmap korelasi antar fitur numerik
numeric_df = df.select_dtypes(include=['number'])  # Memastikan hanya kolom numerik
correlation_matrix = numeric_df.corr().fillna(0)  # Mengisi NaN jika ada

plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.5)
plt.title("Korelasi Antar Fitur Numerik")
plt.show()

# 5. Menangani Outlier dengan Winsorizing
outlier_columns = ["Age", "Financial Stress"]  # Sesuaikan dengan fitur yang memiliki outlier
for col in outlier_columns:
    df[col] = mstats.winsorize(df[col], limits=[0.05, 0.05])  # Perbaikan pemanggilan winsorize

# 6. Menangani Missing Values
# Mengisi NaN dengan median dari masing-masing kolom numerik, hanya untuk kolom numerik
numeric_columns = df.select_dtypes(include=[np.number]).columns
df[numeric_columns] = df[numeric_columns].apply(lambda x: x.fillna(x.median()))

# 7. Mengonversi kolom kategori ke numerik (Encoding)
categorical_columns = df.select_dtypes(include=['object']).columns
le = LabelEncoder()
for col in categorical_columns:
    df[col] = le.fit_transform(df[col])

# 8. Reduksi Dimensi dengan PCA
num_features = df.select_dtypes(include=[np.number]).columns
scaler = StandardScaler()
df_scaled = scaler.fit_transform(df[num_features])

pca = PCA(n_components=2, random_state=42)
df_pca = pca.fit_transform(df_scaled)

# Konversi hasil PCA ke DataFrame
df_pca = pd.DataFrame(df_pca, columns=["PC1", "PC2"])

# 9. Penjelasan Mengapa Preprocessing Ini Diperlukan dan Dampaknya
penjelasan = """
Penjelasan:
A. Penanganan Outlier dengan Winsorizing
   - Mengurangi dampak nilai ekstrem yang dapat mengganggu model.
   - Membantu model lebih stabil dan akurat.

B. Menangani Missing Values dengan Median
   - Menghindari error akibat data kosong saat proses standarisasi dan PCA.
   - Memastikan semua fitur memiliki nilai yang valid.

C. Encoding Kolom Kategori
   - Mengubah data kategori menjadi angka agar dapat digunakan dalam model machine learning.

D. Reduksi Dimensi dengan PCA
   - Mengurangi redundansi informasi akibat korelasi tinggi antar fitur.
   - Mempercepat proses training model dan meningkatkan generalisasi.
"""
print(penjelasan)

# Menampilkan hasil setelah preprocessing
print("\nData setelah PCA:")
print(df_pca.head())

```
